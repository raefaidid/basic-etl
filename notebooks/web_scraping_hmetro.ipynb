# Phase 1: Get the list of articles

```python
import pandas as pd
import requests
```

```python
#169 - utama, 167 - mutakhir
# keys: 'nid', 'title', 'created', 'changed', 'sponsored', 'status', 'url', 'internal_url', 'emotion', 'segment', 'sentiment', 'field_force_ads', 'field_article_author', 'field_author_email', 'field_article_authors', 'field_article_topic', 'field_tags', 'field_article_images', 'field_article_images_filtered', 'field_article_videos', 'field_image_listing_featured_v2', 'field_image_listing_v2', 'field_image_socialmedia', 'field_image_slider', 'field_image_variasi', 'field_image_listing_featured', 'field_image_listing_th', 'field_image_th', 'field_article_state', 'field_article_lead', 'body', 'body_with_inline', 'word_count'
pd.set_option('display.max_colwidth', None)

lst_articles = []

for i in range(10):
    response = requests.get(f"https://www.hmetro.com.my/api/topics/169?page={i}")
    data = response.json()
    lst_articles.extend(data)

dataset = []

for index, item in enumerate(lst_articles):
    
    
    observations = {}
    observations['title'] = item['title']
    observations['url'] = item['url']
    try:
        observations['author_name'] = item['field_article_author']["name"]
    except:
        observations['author_name'] = None
    observations['word_count'] = item['word_count']

    dataset.append(observations)

df = pd.DataFrame(dataset)
display(df)
display(df.info())
    
    
```

```python
df.to_csv("../data/hmetro_articles_without_body.csv", index=False)
```

```python
def get_body(url):
    article_url = f"https://www.hmetro.com.my/{url}"
    response = requests.get(article_url)
    soup = BeautifulSoup(response.text, 'html.parser')
    body = soup.find_all('article-component')
    string = re.search(r"\{.*\}", str(body)).group(0)
    string_parsed_html = html.unescape(string)
    json_string = json.loads(string_parsed_html)
    return json_string['body']
```

```python
df['body'] = df['url'].apply(get_body)
df
```
